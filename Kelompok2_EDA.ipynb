## Load Dataset

```python
df_campak=pd.read_csv("//content//sample_data//dataset//dinkes-od_18508_jumlah_kasus_penyakit_campak_berdasarkan_kabupatenkota_data.csv")
df_tetanus=pd.read_csv("//content//sample_data//dataset//dinkes-od_18514_jml_kasus_penyakit_tetanus__kabupatenkota_data.csv")
df_tbc=pd.read_csv("//content//sample_data//dataset//dinkes-od_18515_jml_kasus_penyakit_tuberkulosis__kabupatenkota_data.csv")
df_diare=pd.read_csv("//content//sample_data//dataset//dinkes-od_18553_jumlah_kasus_penyakit_diare_berdasarkan_kabupatenkota_data.csv")
df_malaria=pd.read_csv("//content//sample_data//dataset//dinkes-od_18554_jml_kasus_penyakit_malaria__kabupatenkota_data.csv")
```

## Drop atribut
Melakukan drop pada atribut 'kode_provinsi', 'nama_provinsi', 'satuan', 'id' karena cakupan dari dataset ini hanya pada provinsi Jawa Barat saja.

```python
df_campak = df_campak.drop(['kode_provinsi', 'nama_provinsi', 'satuan', 'id'],axis=1)
```

```python
df_tetanus = df_tetanus.drop(['kode_provinsi', 'nama_provinsi', 'satuan', 'id'],axis=1)
```

```python
df_tbc = df_tbc.drop(['kode_provinsi', 'nama_provinsi', 'satuan', 'id'],axis=1)
```

```python
df_diare = df_diare.drop(['kode_provinsi', 'nama_provinsi', 'satuan', 'id'],axis=1)
```

```python
df_malaria = df_malaria.drop(['kode_provinsi', 'nama_provinsi', 'satuan', 'id'],axis=1)
```

## Memfilter tiap dataset
Melakukan filterasi agar interval tiap data set sama dari tahun 2018 sampai 2021.

```python
con1 = (df_tetanus['tahun'] >= 2018) & (df_tetanus['tahun'] <= 2021)
df_tetanus = df_tetanus[con1].reset_index(drop=True)
df_tetanus.to_csv('filtered_data.csv', index=False)
```

```python
con1 = (df_tbc['tahun'] >= 2018) & (df_tbc['tahun'] <= 2021)
df_tbc = df_tbc[con1].reset_index(drop=True)
df_tbc.to_csv('filtered_data.csv', index=False)
```

```python
con1 = (df_diare['tahun'] >= 2018) & (df_diare['tahun'] <= 2021)
df_diare = df_diare[con1].reset_index(drop=True)
df_diare.to_csv('filtered_data.csv', index=False)
```

```python
con1 = (df_malaria['tahun'] >= 2018) & (df_malaria['tahun'] <= 2021)
df_malaria = df_malaria[con1].reset_index(drop=True)
df_malaria.to_csv('filtered_data.csv', index=False)
```

## Mengganti data yang tidak masuk akal menjadi mean

##### Mengganti data penyakit diare Kabupaten Tasikmalaya pada tahun 2018

```python
df=pd.read_csv("//content//sample_data//dataset//dinkes-od_18553_jumlah_kasus_penyakit_diare_berdasarkan_kabupatenkota_data.csv")

df[(df['nama_kabupaten_kota'] == 'KABUPATEN TASIKMALAYA')][['nama_kabupaten_kota', 'tahun', 'jumlah_kasus']]
```

Terlihat pada data diatas, jumlah penyakit diare di Kabupaten Tasikmalaya pada tahun 2018 tidak masuk akal jika dibandingkan dengan jumlah kasus diare di tahun sebelum dan setelahnya

```python

data_tasikmalaya = df_merged[(df_merged['nama_kabupaten_kota'] == 'KABUPATEN TASIKMALAYA') & (df_merged['tahun'].isin([2019, 2020, 2021]))]

# Hitung total jumlah_kasus_diare untuk tahun-tahun tersebut
total_kasus_diare = data_tasikmalaya['jumlah_kasus_diare'].mean()

# mengubah menjadi integer, karena satuan disini adalah jiwa/orang
total_kasus_diare = int(total_kasus_diare)

# Update nilai jumlah_kasus_diare untuk tahun 2018 di KABUPATEN TASIKMALAYA
df_merged.loc[(df_merged['nama_kabupaten_kota'] == 'KABUPATEN TASIKMALAYA') & (df_merged['tahun'] == 2018), 'jumlah_kasus_diare'] = total_kasus_diare
```

##### Mengganti data penyakit diare Kabupaten Indramayu pada tahun 2018

```python

df[(df['nama_kabupaten_kota'] == 'KABUPATEN INDRAMAYU')][['nama_kabupaten_kota', 'tahun', 'jumlah_kasus']]
```

Terlihat pada data diatas, jumlah penyakit diare di Kabupaten Indramayu pada tahun 2018 tidak masuk akal dan kami anggap tidak terdata dikarenakan tidak mungkin jumlah kasus diare bisa 0 kalau jumlah kasus beberapa tahun sebelum dan sesudahnya konsisten di 40.000, juga apabila dibandingkan dengan provinsi lain di tahun yang sama juga, hanya Kabupaten Indramayu yang jumlah kasus diarenya 0

```python

data_indramayu = df_merged[(df_merged['nama_kabupaten_kota'] == 'KABUPATEN INDRAMAYU') & (df_merged['tahun'].isin([2019, 2020, 2021]))]

# Hitung total jumlah_kasus_diare untuk tahun-tahun tersebut
total_kasus_diare = data_indramayu['jumlah_kasus_diare'].mean()

# mengubah menjadi integer, karena satuan disini adalah jiwa/orang
total_kasus_diare = int(total_kasus_diare)

# Update nilai jumlah_kasus_diare untuk tahun 2018 di KABUPATEN INDRAMAYU
df_merged.loc[(df_merged['nama_kabupaten_kota'] == 'KABUPATEN INDRAMAYU') & (df_merged['tahun'] == 2018), 'jumlah_kasus_diare'] = total_kasus_diare
```

##### Mengganti data penyakit diare Kabupaten Sumedang pada tahun 2018

```python


df[(df['nama_kabupaten_kota'] == 'KABUPATEN SUMEDANG')][['nama_kabupaten_kota', 'tahun', 'jumlah_kasus']]
```

Terlihat pada data diatas, jumlah penyakit diare di Kabupaten Sumedang pada tahun 2018 tidak masuk akal jika dibandingkan dengan jumlah kasus diare di tahun sebelum dan setelahnya

```python

data_sumedang = df_merged[(df_merged['nama_kabupaten_kota'] == 'KABUPATEN SUMEDANG') & (df_merged['tahun'].isin([2019, 2020, 2021]))]

# Hitung total jumlah_kasus_diare untuk tahun-tahun tersebut
total_kasus_diare = data_sumedang['jumlah_kasus_diare'].mean()

# mengubah menjadi integer, karena satuan disini adalah jiwa/orang
total_kasus_diare = int(total_kasus_diare)

# Update nilai jumlah_kasus_diare untuk tahun 2018 di KABUPATEN SUMEDANG
df_merged.loc[(df_merged['nama_kabupaten_kota'] == 'KABUPATEN SUMEDANG') & (df_merged['tahun'] == 2018), 'jumlah_kasus_diare'] = total_kasus_diare
```

## Heat Map untuk melihat korelasi antar penyakit


```python
numeric_cols = ['jumlah_kasus_tetanus', 'jumlah_kasus_campak', 'jumlah_kasus_tbc', 'jumlah_kasus_diare', 'jumlah_kasus_malaria']
heat_data = df_merged[numeric_cols].corr()
sns.heatmap(heat_data, annot=True)
plt.show()
```

Tampak bahwa korelasi antara jumlah kasus TBC dan jumlah kasus diare cukup tinggi, dengan nilai 0.8. Ini menunjukkan adanya korelasi positif yang kuat antara jumlah kasus TBC dan jumlah kasus diare.

1.   Mungkin ada faktor demografis atau populasi yang memengaruhi kedua jenis penyakit ini. Misalnya, jika wilayah tertentu memiliki populasi yang lebih rentan terhadap penyakit infeksi, baik TBC maupun diare, maka jumlah kasus keduanya dapat meningkat secara bersamaan.
2.   Faktor lingkungan tertentu, seperti sanitasi yang buruk atau kualitas air yang rendah, dapat meningkatkan risiko terkena baik TBC maupun diare. Jika lingkungan memiliki peran dalam penyebaran kedua penyakit ini, maka korelasi yang tinggi bisa terjadi.
3.   Jika ada kesamaan dalam metode penularan atau rute penyebaran penyakit antara TBC dan diare, maka hal itu bisa menghasilkan korelasi yang tinggi.
